```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from sklearn.datasets import make_classification
```

```python
X,Y = make_classification(n_samples=1000,n_features=2,n_informative=2,n_repeated=0,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=0)
```

```python
print(X.shape)
print(Y.shape)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()
```

```python
x = np.array(X)
y = np.array(Y) 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
print(len(x_test))
print(len(x_train))
```

```python
def hypothesis(x,w,b):
    h = np.dot(x,w)+b
    return sigmoid(h)

def sigmoid(z):
    return (1.0)/(1.0 + np.exp(-z))

def error(x,y,w,b):
    m = x.shape[0]
    err = 0.0
    
    for i in range(m):
        y_pred = hypothesis(x[i],w,b)
        err = err + (y[i]*np.log2(y_pred)) + ((1-y[i])*np.log2(1-y_pred)) 
        
    return -err/m 

def gradient(x,y,w,b):
    grad_w = np.zeros(x.shape[1])
    grad_b = 0.0
    
    m=x.shape[0]
    
    for i in range(m):
        y_pred = hypothesis(x[i],w,b)
        grad_w += (y[i]-y_pred)*x[i]
        grad_b += (y[i]-y_pred)
        
    grad_w = grad_w/m
    grad_b = grad_b/m
        
    return [grad_w,grad_b]

def grad_descent(x,y,lr=0.5,epochs=500):
    loss=[]
    w = np.zeros(x.shape[1])
    b = 0.0
    
    for i in range(epochs):
        y_pred = hypothesis(x[i],w,b)
        e = error(x,y,w,b)
        [grad_w , grad_b] = gradient(x,y,w,b)
        
        w = w + lr*grad_w
        b = b + lr*grad_b
        
        loss.append(e)
        print("Epoch: "+str(i) + " Cost: "+str(e) + " Weight: "+str(w) + " Bias: "+str(b))
        
    return loss,w,b
```

```python
L,W,b = grad_descent(x_train,y_train)
```

```python
plt.plot(L)
plt.show()
```

```python
final=[]
for i in range(x_test.shape[0]):
    confidence = hypothesis(x_test[i],W,b)
    if(confidence<0.5):
        final.append(0)
    else:
        final.append(1)
```
