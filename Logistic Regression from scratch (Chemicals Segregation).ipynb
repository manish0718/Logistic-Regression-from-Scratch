```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x_train = pd.read_csv('Logistic_X_Train.csv')
x_test = pd.read_csv('Logistic_X_Test.csv')
y_train = pd.read_csv('Logistic_Y_Train.csv')

#Converting pandas data frame to numpy array 
x_train = np.array(x_train)
x_test = np.array(x_test)
y_train = np.array(y_train)

print(x_test.shape)
```

```python
def hypothesis(x,w,b):
    h = np.dot(x,w)+b
    return sigmoid(h)

def sigmoid(z):
    return (1.0)/(1.0 + np.exp(-z))

def error(x,y,w,b):
    m = x.shape[0]
    err = 0.0
    
    for i in range(m):
        y_pred = hypothesis(x[i],w,b)
        err = err + (y[i]*np.log2(y_pred)) + ((1-y[i])*np.log2(1-y_pred)) 
        
    return -err/m 

def gradient(x,y,w,b):
    grad_w = np.zeros(x.shape[1])
    grad_b = 0.0
    
    m=x.shape[0]
    
    for i in range(m):
        y_pred = hypothesis(x[i],w,b)
        grad_w += (y[i]-y_pred)*x[i]
        grad_b += (y[i]-y_pred)
        
    grad_w = grad_w/m
    grad_b = grad_b/m
        
    return [grad_w,grad_b]

def grad_descent(x,y,lr=0.1,epochs=1000):
    loss=[]
    w = np.zeros(x.shape[1])
    b = 0.0
    
    for i in range(epochs):
        y_pred = hypothesis(x[i],w,b)
        e = error(x,y,w,b)
        [grad_w , grad_b] = gradient(x,y,w,b)
        
        w = w + lr*grad_w
        b = b + lr*grad_b
        
        loss.append(e)
        print("Epoch: "+str(i) + " Cost: "+str(e) + " Weight: "+str(w) + " Bias: "+str(b))
        
    return loss,w,b
```

```python
L,W,b = grad_descent(x_train,y_train)
```

```python
plt.plot(L)
plt.show()
```

```python
final=[]
for i in range(x_test.shape[0]):
    confidence = hypothesis(x_test[i],W,b)
    if(confidence<0.5):
        final.append(0)
    else:
        final.append(1)
```
